```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df["show_id"]=df["show_id"].str.lower()
df["type"]=df["type"].str.lower()
df["title"]=df["title"].str.lower()
df["director"]=df["director"].str.lower()
df["cast"]=df["cast"].str.lower()
df["country"]=df["country"].str.lower()
df["rating"]=df["rating"].str.lower()
df["listed_in"]=df["listed_in"].str.lower()
df["description"]=df["description"].str.lower()
```

```python
df.head(3)
```

```python
df["show_id"]=df["show_id"].str.strip()
df["type"]=df["type"].str.strip()
df["title"]=df["title"].str.strip()
df["director"]=df["director"].str.strip()
df["cast"]=df["cast"].str.strip()
df["country"]=df["country"].str.strip()
df["rating"]=df["rating"].str.strip()
df["listed_in"]=df["listed_in"].str.strip()
df["description"]=df["description"].str.strip()
```

```python
df.head(2)
```

```python
df.rename(columns={"type":"types","cast":"casts"})
```

```python
#dropping_duplicates
df.drop_duplicates(inplace=True)
```

```python
df["added_date"]=pd.to_datetime(df["date_added"],format="%B %d, %Y",errors='coerce')
```

```python
df.info()
```

```python
# Find rows where added_date is NaT but date_added is not NaN
unparsed_dates = df[df['added_date'].isna() & df['date_added'].notna()]

# Display some of the original date_added values that were not parsed
if not unparsed_dates.empty:
    print("Sample of date_added values that were not parsed:")
    display(unparsed_dates['date_added'].sample(min(10, len(unparsed_dates))))
else:
    print("All non-null date_added values were successfully parsed.")
```

```python
df['date_added'] = df['date_added'].str.strip()
df["added_date"] = pd.to_datetime(df["date_added"], format="%B %d, %Y", errors='coerce')

# Check how many values are still NaT after stripping whitespace
print(f"Number of NaT values in 'added_date' after stripping whitespace: {df['added_date'].isna().sum()}")

# Display info to see the non-null counts
display(df.info())
```

```python
df.head()
df.info()
```

```python
df.info()
```

```python
# Calculate the number of null values in each row
null_counts_per_row = df.isnull().sum(axis=1)
# print(null_counts_per_row)

# Identify rows with more than 5 null values
rows_to_drop = null_counts_per_row[null_counts_per_row > 3].index
#print(rows_to_drop)

# Drop the identified rows
df_cleaned = df.drop(rows_to_drop)

# Display the shape of the original and cleaned DataFrame to show the number of rows dropped
print(f"Original DataFrame shape: {df.shape}")
print(f"DataFrame shape after dropping rows with more than 5 null values: {df_cleaned.shape}")
```

```python
df_cleaned.rename(columns={"type":"types","cast":"casts"},inplace=True)
```

```python
df_cleaned.info()
```

```python
df_cleaned['director'].fillna('Unknown', inplace=True)

# Checking the number of nulls in the 'director' column after filling
print(f"Number of null values in 'director' column after filling: {df_cleaned['director'].isnull().sum()}")
```

```python
# do the same for the cast and country
df_cleaned['casts'].fillna('Unknown', inplace=True)
df_cleaned['country'].fillna('Unknown', inplace=True)
```

```python
#dropping rows with null values in  rating and duration
df_cleaned.dropna(subset=["rating","duration"],inplace=True)
```
